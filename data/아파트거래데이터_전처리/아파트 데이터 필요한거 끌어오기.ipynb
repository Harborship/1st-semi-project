```python
import pandas as pd
from tqdm import tqdm
```

```python
apt_df=pd.read_csv("아파트_리스트_kr.csv", encoding='cp949')
df=pd.read_csv("전세_전처리_데이터.csv")
```

```python
df.head()
```

```python
apt_df.head()
```

```python
data_lst=[]
for i in df.index:
    data_lst.append(  df.loc[i, '주소'].split(" ")[-2]+" "+df.loc[i, '단지명']  )
    
df['동_단지']=data_lst
```

```python
data=list(df['동_단지'].unique())
data
```

```python
index_lst=[]

for i in apt_df.index:
    if apt_df.loc[i, 'apt'] in data:
        index_lst.append(i)
        
index_lst
```

```python
new_apt=apt_df.loc[index_lst, :]
new_apt.reset_index(drop=True, inplace=True)
```

```python
new_apt.to_csv('전세_아파트_리스트.csv', index=False, encoding='cp949')
```
