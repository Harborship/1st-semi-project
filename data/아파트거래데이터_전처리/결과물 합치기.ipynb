```python
import pandas as pd
```

```python
df1=pd.read_csv("./0-1400.csv")
df1.drop('Unnamed: 0', axis=1, inplace=True)

df2=pd.read_csv("./1400_2800.csv")
df2.drop('Unnamed: 0', axis=1, inplace=True)

df3=pd.read_csv("./2800-4159.csv")

df4=pd.read_csv("./전세_아파트_리스트.csv", encoding='cp949')

df5=pd.read_csv("./차집합데이터.csv")
df5.drop('Unnamed: 0', axis=1, inplace=True)
```

```python
df_total=pd.concat([df1,df2,df3,df5])
df_total=df_total[df_total['new_zibun']!='None']
df_total.reset_index(inplace=True, drop=True)
```

```python
df_drop = df_total.drop_duplicates(['apt', 'new_road','home_cnt','home_size','room_cnt', 'bath_cnt'])
df_drop=df_drop[['apt', 'road', 'new_apt', 'new_road', 'latitude', 'longitude',
       'company', 'home_cnt', 'home_size', 'room_cnt', 'bath_cnt', 'bigo',
       'zibun', 'new_zibun']]
df_drop
```

```python
df4
```

```python
df_result=pd.concat([df4, df_drop])
df_result = df_result.drop_duplicates(['apt', 'new_road', 'home_cnt', 'home_size', 'room_cnt', 'bath_cnt'])
df_result=df_result[df_result['latitude']!=0]
df_result=df_result[df_result['bath_cnt']!='-']
df_result.reset_index(drop=True, inplace=True)
df_result
```

```python
df_result.to_csv('./전세_아파트_리스트_완성본.csv', index=False, encoding='cp949')
```

```python
df=pd.read_csv("./전세_전처리_데이터.csv")
df
```

```python
df_final=pd.merge(df, df_result, left_on=['동_단지', '전용면적(㎡)'], right_on=['apt', 'home_size'], how='inner')
df_final = df_final.drop_duplicates(['주소','전용면적(㎡)','계약년월','계약일','보증금(만원)'])
df_final
```

```python
df_final.to_csv("./전세_실거래_데이터_완성본.csv", encoding='cp949')
```
