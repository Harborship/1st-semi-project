```python
import re
import pandas as pd
```

```python
df=pd.read_csv("./data/서울시 주요 공원현황.csv", encoding='cp949')
df.dropna(subset='면적', inplace=True)
df=df[['면적', 'X좌표(WGS84)', 'Y좌표(WGS84)']]
df_result=df.drop([6])
df_result['면적']=df_result['면적'].apply(lambda x : x.split("㎡")[0])
df_result['면적']=df_result['면적'].apply(lambda x : re.sub("[^0-9]", "", x))
df_result=df_result.astype({"면적" : float})
df_result.columns=["면적", '경도', '위도']
df_result
```

```python
df_result.to_csv("./data/서울시 주요 공원현황_전처리.csv", encoding='cp949')
```
